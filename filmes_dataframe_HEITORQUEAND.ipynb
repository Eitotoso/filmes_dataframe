# QUESTÃO 1

```python
import pandas as pd

dados = {
    'Nome': ['Vingadores: Ultimato', 'O Senhor dos Anéis: O Retorno do Rei', 'Que Horas Ela Volta', 'A Viagem de Chihiro', 'O Poderoso Chefão', '1917', 'Avatar'],
    'Diretor': ['Anthony e Joe Russo', 'Peter Jackson', 'Anna Muylaert', 'Hayao Miyazaki', 'Francis Ford Coppola', 'Sam Mendes', 'James Cameron'],
    'Gênero': ['Ação', 'Fantasia', 'Drama', 'Animação', 'Drama', 'Guerra', 'Ação/Ficção científica'],
    'Duração em horas': [3, 3.2, 1.5, 2.5, 2.5, 1.5, 2.4],
    'Ator principal': ['Robert Downey Jr.', 'Elijah Wood', 'Regina Casé', 'Rumi Hiiragi', 'Marlon Brando', 'George MacKay', 'Sam Worthington'],
    'Nota do IMDb': [8.4, 8.9, 7.7, 8.6, 9.2, 8.3, 7.9]
}

filmes_df = pd.DataFrame(dados)
display (filmes_df)
```

# QUESTÃO 2

```python
# Apresentando o tamanho:
tamanho = filmes_df.shape
print(f"Tamanho do DataFrame: {tamanho[0]} linhas x {tamanho[1]} colunas")
```

# QUESTÃO 7

```python
# Adicionando novo filme
novoFilme = {
    'Nome': 'O Homem do Futuro',
    'Diretor': 'Cláudio Torres',
    'Gênero': 'Comédia',
    'Duração em horas': 1.5,
    'Ator principal': 'Wagner Moura',
    'Nota do IMDb': 7.0
}

novoFilme_df = pd.DataFrame([novoFilme])

filmes_df = pd.concat([filmes_df, novoFilme_df], ignore_index=True)

display(filmes_df)
```
